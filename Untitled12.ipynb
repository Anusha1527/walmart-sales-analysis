```python
import pandas as pd

df = pd.read_csv("Walmart DataSet.csv")
df.head()
```

```python
df.columns
```

```python
#Convert Date Column
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
df['Date'].head()
```

1. Data Preprocessing & EDA

a) Handling Missing Values

```python
df.isnull().sum()
df.fillna(df.mean(numeric_only=True), inplace=True)
```

Missing numerical values were handled using mean imputation to preserve overall trends without removing records.

b) Outlier Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['Weekly_Sales'])
plt.show()
```

Outliers are present due to holiday sales and promotions. These values represent real business scenarios and were not removed

1(a). Effect of Unemployment on Weekly Sales

```python
sns.scatterplot(x='Unemployment', y='Weekly_Sales', data=df)
plt.show()
```

```python
store_unemp_corr = df.groupby('Store').apply(
    lambda x: x['Weekly_Sales'].corr(x['Unemployment'])
).sort_values()
store_unemp_corr.head()
```

Weekly sales show a negative correlation with unemployment.

Stores in regions with higher unemployment experience reduced sales.

These stores are the most affected during economic slowdowns.

1(b). Seasonal Trend in Weekly Sales

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month

monthly_sales = df.groupby('Month')['Weekly_Sales'].mean()
monthly_sales.plot(kind='line')
plt.show()
```

Strong seasonality observed.

Peak sales during NovemberDecember due to festive season and holidays.

Lower sales in JanuaryFebruary due to post-holiday slowdown.

1(c). Impact of Temperature on Weekly Sales

```python
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=df)
plt.show()
```

No strong linear relationship between temperature and sales.

Moderate temperatures show slightly higher sales.

Extreme weather conditions reduce customer footfall.

1(d). Effect of Consumer Price Index (CPI)

```python
sns.scatterplot(x='CPI', y='Weekly_Sales', data=df)
plt.show()
```

Increase in CPI slightly reduces weekly sales.

Higher CPI lowers consumer purchasing power, impacting demand.

1(e). Top Performing Stores

```python
top_stores = df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False)
top_stores.head()
```

Top-performing stores consistently generate high sales.

Likely due to high population density, location advantage, and stable demand.

1(f). Worst Performing Store & Performance Gap

```python
best_store_sales = top_stores.max()
worst_store_sales = top_stores.min()
difference = best_store_sales - worst_store_sales
difference
```

Significant gap between the highest and lowest performing stores.

Indicates uneven regional demand and economic conditions.

2. Sales Forecasting (Next 12 Weeks)
Technique Used

Time Series Forecasting using ARIMA

Forecasting for One Store (Repeatable for All Stores)

```python
from statsmodels.tsa.arima.model import ARIMA

store1 = df[df['Store'] == 1]
store1_ts = store1.groupby('Date')['Weekly_Sales'].sum()

model = ARIMA(store1_ts, order=(1,1,1))
model_fit = model.fit()

forecast = model_fit.forecast(steps=12)
forecast
```

Visualization

```python
plt.figure(figsize=(10,5))
plt.plot(store1_ts, label='Historical Sales')
plt.plot(forecast, label='Next 12 Weeks Forecast', color='red')
plt.legend()
plt.show()
```

The model predicts future sales based on historical trends.

Forecasting helps Walmart plan inventory efficiently and reduce over-stocking or shortages.
